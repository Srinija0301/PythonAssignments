**Give rank to country by continent based on life expectancy, gdp and life expectancy and gdp ratio.**

```python
df['le_gdp_ratio'] = df['lifeExp']/df['gdpPercap']
```

```python
df.head()
```

```python
sorted = df.sort_values(by=['lifeExp', 'gdpPercap','le_gdp_ratio'])
```

```python
df['country_rank']=sorted.groupby(['continent'])['year'].rank()
df.sort_values('country_rank')
```

**b) Give rank to continent by life expectancy, gdp and life expectancy and gdp ratio.**

```python
df1= df.sort_values(['lifeExp','gdpPercap','le_gdp_ratio'])
df['continent_rank'] =  df1.groupby(['year','continent'])['continent'].rank(method="average")
df.sort_values('continent_rank')
```

**Give rank to country, continent by life expectancy by year.**


```python
df1 = df.sort_values(["lifeExp","year","continent"])
df['country_by_year']=df1.groupby(['lifeExp','year','continent'])['country'].rank()
df
```
