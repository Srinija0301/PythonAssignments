```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('adult.data.csv')
```

```python
df
```

```python
#no NAN data
df[df.isna().any(axis=1)]
```

```python
print("No of Male   :", len(df[df["sex"]=="Male"]))
print("No of Female :",len(df[df["sex"]=="Female"]))
```

```python
print("Average age of woman :",np.average(df[df["sex"]=="Female"]["age"]))
```

```python
print("Percentage of German Citizens :", (100*len(df[df["native-country"]=="Germany"]))/len(df),"%")
```

```python
print("For people who earn >50, mean of age :",np.average(df[df["salary"]==">50K"]["age"]),",standard deviation of age :",np.std(df[df["salary"]==">50K"]["age"]))
print("For people who earn <=50, mean of age :",np.average(df[df["salary"]=="<=50K"]["age"]),",standard deviation of age :",np.std(df[df["salary"]=="<=50K"]["age"]))
```

```python
#False, people other than (Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate) education also earn >50K salary
df[df["salary"]==">50K"]["education"].unique().tolist()
```

```python
for index,group in df.groupby(["sex","race"]):
  print(index)
  print(group[["age"]].describe())
  print("---------------------------------")
print("Max age of men of Amer-Indian-Eskimo is 82")
```

```python
married_columns = [i for i in df["marital-status"].unique().tolist() if i.find("Married")==(0) or i.find("married")==(0)  ]
married_columns
```

```python
print("people who earn >50K and are Married :",len(df[(df["salary"]==">50K") & (df["marital-status"].isin(married_columns))]))
print("people who earn >50K and are Bachelors :",len(df[(df["salary"]==">50K") & (~df["marital-status"].isin(married_columns))]))
```

```python
print("maximum number of hours a person works per week :",max(df["hours-per-week"]))
print("people working maximum number of hours :", len(df[df["hours-per-week"]==max(df["hours-per-week"])]))
print("number of people working maximum number of hours and earn >50K :", len(df[(df["hours-per-week"]==max(df["hours-per-week"])) & (df["salary"]==">50K")]))
print("% of people working maximum number of hours and earning >50K :", (100*len(df[(df["hours-per-week"]==max(df["hours-per-week"])) & (df["salary"]==">50K")]))/len(df[df["hours-per-week"]==max(df["hours-per-week"])]))
```

```python
for index,group in df.groupby(["salary","native-country"]):
  print(index)
  print(group[["hours-per-week"]].describe())
  print("---------------------------------")
print("Average hours-per-week for people with >50K salary in Japan : 47.958333")
print("Average hours-per-week for people with <=50K salary in Japan : 41.000000")
```
